```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data

X_train=train_df.drop(columns= ["index","loan_status","Unnamed: 0"])
X_train = pd.get_dummies(X_train)
y_train=LabelEncoder().fit_transform(train_df["loan_status"])
y_train

train_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for testing data

X_test= test_df.drop(columns=["index","loan_status","Unnamed: 0"])
X_test= pd.get_dummies(X_test)
y_test = LabelEncoder().fit_transform(test_df["loan_status"])
X_test['debt_settlement_flag_Y'] = 0

test_df.head()
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score

classifier = LogisticRegression(random_state=1).fit(X_train, y_train)
print(f"Logistic Regression Training Score: {classifier.score(X_train, y_train)}")
print(f"Logistic Regression Testing Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"10 predictions:   {predictions[:10]}")
print(f"10 actual labels: {y_test[:10].tolist()}")
```

```python
y_true = y_test
y_pred = classifier.predict(X_test)
confusion_matrix(y_true, y_pred)
```

```python
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
accuracy = (tp + tn) / (tp + fp + tn + fn) 
print(f"Accuracy: {accuracy}")
```

```python
print(classification_report(y_test, predictions, target_names=["low_risk","high_risk"]))
```

```python
# Train a Random Forest Classifier model and print the model score

classifier2 = RandomForestClassifier(random_state=1).fit(X_train, y_train)
print(f'Random Forest Classifier Training Score: {classifier2.score(X_train, y_train)}')
print(f'Random Forest Classifier Testing Score: {classifier2.score(X_test, y_test)}')
```

```python
y_pred = classifier2.predict(X_test)
print(classification_report(y_test, y_pred, target_names=["low_risk","high_risk"]))
```

```python
# Scale the data

scaled_data = StandardScaler().fit(X_train)
X_train_scaled = scaled_data.transform(X_train)
X_test_scaled = scaled_data.transform(X_test)
classifier.fit(X_train_scaled, y_train)
print(f"Scaled Log Training Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Scaled Log Testing Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
predictions = classifier.predict(X_test_scaled)
print(f"10 predictions:   {predictions[:10]}")
print(f"10 actual labels: {y_test[:10].tolist()}")
```

```python
y_true = y_test
y_pred = classifier.predict(X_test_scaled)
confusion_matrix(y_true, y_pred)
```

```python
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
accuracy = (tp + tn) / (tp + fp + tn + fn) 
print(f"Accuracy: {accuracy}")
```

```python
print(classification_report(y_test, predictions, target_names=["low_risk","high_risk"]))
```

```python
feature_importances = classifier2.feature_importances_
features = sorted(zip(X_test.columns, classifier2.feature_importances_), key = lambda x: x[1])
cols = [f[0] for f in features]
width = [f[1] for f in features]
fig, ax = plt.subplots()
fig.set_size_inches(10,200)
plt.margins(y=0.001)
ax.barh(y=cols, width=width)

plt.show()
```
